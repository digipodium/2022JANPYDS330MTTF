# slicing in numpy

````python
import numpy as np
````

````python
x= np.random.randint(1,10, (5,5))
print(x)
````

selecting a row

````python
x[0] # x[0,:]
````

single value (item) 

````python
x[0][3]
````

single value (item) coordinate style

````python
x[0,3]
````

````python
x[:,0]
````

````python
x[:, -1] # last column
````

````python
x[1:,:-1]
````

````python
x = np.random.randint(1,20,(2,2))
y = np.random.randint(2,100,(3,2))
print(x)
print(y)
````

````python
np.vstack((x,y))
````

````python
yt = y.transpose()
yt.shape
````

````python
np.hstack((x,yt))
````

````python
x = np.arange(-10,10,.02)
y = np.sin(x)
z = np.cos(x)
````

````python
import plotly.express as px
import matplotlib.pyplot as plt
````

````python
fig = px.scatter_3d(x=x,y=y,z=z,height=500,width=500,color=z,)
fig.show() 
# fig.write_html('graph1.html',auto_open=True) # if any error happens
````
